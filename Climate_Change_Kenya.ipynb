## Preparing The DataSets

The datasets we have imported have our expected column headers on a row in the index so we will set the correct headers and remove the metadata included as column headers.

We will also filter out data from other countries to remain only with data from Kenya since we are interested in Kenya then combine the 3 dataFrames to one to be used for analysis.

1. Prevalence of undernourishment (% of population) Data

```python
# setting column header
Undernourished_pop = pd.read_csv('/content/drive/MyDrive/UNBDH/Prevalence of undernourishment (% of population).xlsx - Data.csv', header = [3])

# previewing dataFrame
Undernourished_pop.head()
```

```python
# filtering out data from other countries to remain only with data from Kenya
Undernourished_pop_Kenya = Undernourished_pop[Undernourished_pop['Country Name']== 'Kenya']

# previewing dataFrame
Undernourished_pop_Kenya
```

2. Prevalence of severe food insecurity in the population (%) Data



```python
# setting column header
Food_insecure_pop = pd.read_csv('/content/drive/MyDrive/UNBDH/Prevalence of severe food insecurity in the population (%).xlsx - Data.csv', header = [3])

# previewing dataFrame
Food_insecure_pop.head()
```

```python
# filtering out data from other countries to remain only with data from Kenya
Food_insecure_pop_Kenya = Food_insecure_pop[Food_insecure_pop['Country Name']== 'Kenya']

# previewing dataFrame
Food_insecure_pop_Kenya
```

3. Climate Change Data

```python
# setting column header
Climate_change = pd.read_excel('/content/drive/MyDrive/UNBDH/World Climate Change.xlsx', header = [4])

# previewing dataFrame
Climate_change.head()
```

```python
# filtering out data from other countries to remain only with data from Kenya
Climate_change_Kenya = Climate_change[Climate_change['Country Name']== 'Kenya']

# previewing dataFrame
Climate_change_Kenya
```

```python
# combining the 3 dataFrames to one

dataFrames = [Climate_change_Kenya, Undernourished_pop_Kenya, Food_insecure_pop_Kenya]

data = pd.concat(dataFrames)

# previewing dataFrame
data
```

Since all the data we have is from Kenya we can now drop the country name and country code columns and the Indicator code column that we may not need.

```python
# dropping unnecessary columns
data = data.drop(['Country Name', 'Country Code', 'Indicator Code'], axis = 1)

# previewing dataFrame
data
```

```python
# formatting the Indicator Name column for ease during analysis
data.rename(columns = {'Indicator Name':'Indicator_Name'}, inplace = True)

# preview data Frame
data
```

```python
# reindexing and swapping our rows and columns for ease during analysis 
data = data.set_index('Indicator_Name')
data = data.transpose()
data
```

```python
# replacing missing values with zeros
data = data.fillna(0)
data
```

```python
# viewing columns, checking the indicators we have
data.columns
```

```python
# filtering out the other unwanted columns to remain only with the ones we need
data = data.drop(data.columns.difference(['Total greenhouse gas emissions (% change from 1990)', 'Agricultural land (% of land area)', 'Cereal yield (kg per hectare)', 'Average precipitation in depth (mm per year)', 'Agriculture, forestry, and fishing, value added (% of GDP)', 'Prevalence of severe food insecurity in the population (%)', 'Prevalence of undernourishment (% of population)']), 1)

# previewing dataFrame
data
```

## Univariate Analysis

```python
sns.lineplot(data = data, x = data.index, y = 'Total greenhouse gas emissions (% change from 1990)').set(title='Total greenhouse gas emissions (% change from 1990)');
```

*OBSERVATIONS:*
> The total percentage of greenhouse gases emitted was constant before 1990 but then dropped beyond normal in the year 1990 and started rising and dropping gradually.

## Bivariate Analysis

> This means that climate change has no major effect on food production.

## Conclusion and Recommendations

The major finding from the heatmap shows that climate change has very little effect on food production, although it is not completely zero. It has some effect.

3. AsseSsing fuel use in the farm and industries to come up with energy saving methods.
4. PreferRing use of renewable energy like solar, wind and renewable energy where possible.
5. Creating awareness on climate change and advice Kenyan citizens on what to how to tackle climate change, not only in food production but also in all sectors.
